# **1. More unknowns than equations → non-zero null-space vectors**

If (A) is $m \times n$ with (n > m), then (Ax=0) has **more unknowns than equations**.
Thus the system has **free variables**, meaning the null space has **non-zero solutions**.

### **Example (Python)**

Matrix with 2 equations and 3 unknowns:

$$
A=\begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6
\end{bmatrix}
$$

```python
import sympy as sp

A = sp.Matrix([[1,2,3],
               [4,5,6]])
N = A.nullspace()
N
# Output will show **a non-zero vector** (null-space basis).
```

# **2. Linear Independence**

Vectors $x_1,\ldots,x_n$ are independent if:

$$
c_1x_1 + \cdots + c_nx_n = 0 \quad\Rightarrow\quad c_1=\cdots=c_n=0.
$$

### **Example of independent vectors in $\mathbb{R}^2$**

Take:

$$
v_1 = \begin{bmatrix}1 \\ 0\end{bmatrix}, \quad
v_2 = \begin{bmatrix}1 \\ 2\end{bmatrix}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

v1 = np.array([1,0])
v2 = np.array([1,2])

plt.figure()
plt.axhline(0)
plt.axvline(0)

plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1)
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1)

plt.xlim(-1,3)
plt.ylim(-1,3)
plt.gca().set_aspect('equal')
plt.show()
```

These vectors are **not multiples** of each other → independent.
### **Why is the zero vector never independent with another vector?**

Because:

$$
c_1(0) + c_2 v = 0
$$

You can choose $c_2 = 0$ and **any** $c_1$ and the combination still equals 0.
So independence fails (coefficients are not uniquely zero).

# **3. Independence ↔ Null space contains only zero**

Put the vectors into a matrix $A=[v_1\ v_2\ \cdots\ v_n]$.
Then:

* If $\text{Null}(A)={0}$ → columns are **independent**
* If rank (=n) → columns are **independent**
* If rank (<n) → columns are **dependent**

```python
A = sp.Matrix([[1,0],
               [1,2]])

A.rank(), A.nullspace()
# Rank = 2 → null space only {0} → independent.
```

# **4. Span of vectors**

Vectors $v_1,\ldots,v_\ell$ **span a space** when every vector in the space can be formed as a combination:

$$
c_1 v_1 + \cdots + c_\ell v_\ell
$$

# **5. Basis of a vector space**

A basis must satisfy:

1. **Independence**
2. **Span the whole space**

### **Example: Standard basis of (\mathbb{R}^3)**

$$
e_1=[1,0,0],\quad e_2=[0,1,0],\quad e_3=[0,0,1]
$$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

e1 = np.array([1,0,0])
e2 = np.array([0,1,0])
e3 = np.array([0,0,1])

ax.quiver(0,0,0, *e1)
ax.quiver(0,0,0, *e2)
ax.quiver(0,0,0, *e3)

ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.set_zlim([0,1])
plt.show()
```

# **9. Rank & column space dimension**

For matrix (A):

$$
\text{rank}(A) = \dim C(A)
$$

If (A) is $4 \times 4$ and rank = 2:

* Column space dimension = 2
* Any basis must pick **two independent combinations** of columns

Example basis (using Python to find pivot columns):

```python
A = sp.Matrix([[1,2,3,4],
               [2,4,6,8],
               [1,1,1,1],
               [0,1,2,3]])

A.rank(), A.columnspace()
```

- Columnspace() gives one possible basis, but **any two independent vectors in the column space** form a basis.

# **10. Dimension of null space**

$$
\dim N(A) = n - r
$$

where

* (n) = number of columns
* (r) = rank

### Example

If (A) is (4 \times 4) with rank 2:

$$
\dim N(A) = 4 - 2 = 2
$$

Compute with Python:

```python
A.nullspace()
```
